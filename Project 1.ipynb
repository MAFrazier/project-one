```python
import pandas as pd
import numpy as np

# Assuming you have loaded your data into a DataFrame (smoke_data_ct)
smoke_data_ct = pd.read_csv("Behavioral_Risk_Factor_Data__Tobacco_Use__2017-Connecticut.csv")

selected_columns = ['LocationDesc', 'TopicDesc', 'MeasureDesc', 'Response', 'Sample_Size', 'Data_Value']
df_new = smoke_data_ct[selected_columns]

# Drop rows where 'Sample_Size' is NaN after conversion
df_cleaned = df_new.dropna(subset=['Sample_Size'])

# Convert 'Sample_Size' and 'Data_Value' columns to integers
df_cleaned.loc[:, ['Sample_Size', 'Data_Value']] = df_cleaned.loc[:, ['Sample_Size', 'Data_Value']].astype(int)

values_column = df_cleaned['Data_Value']  # Replace 'Values' with your actual column name
weights_column = df_cleaned['Sample_Size']  # Replace 'Weights' with your actual column name

# Calculate the weighted average
weighted_avg = np.average(values_column, weights=weights_column)
median_value = df_cleaned['Data_Value'].median()
range_value = df_cleaned['Data_Value'].max() - df_cleaned['Data_Value'].min()

df_summary = {
    'Weighted_Average': [weighted_avg],
    'Median': [median_value],
    'Range': [range_value]
}

print(df_summary)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming you have loaded your data into a DataFrame (smoke_data_ct)
smoke_data_ct = pd.read_csv("Behavioral_Risk_Factor_Data__Tobacco_Use__2017-Connecticut.csv")

selected_columns = ['LocationDesc', 'TopicDesc', 'MeasureDesc', 'Response', 'Sample_Size', 'Data_Value']
df_new = smoke_data_ct[selected_columns]

# Drop rows where 'Sample_Size' is NaN after conversion
df_cleaned = df_new.dropna(subset=['Sample_Size'])

# Convert 'Sample_Size' and 'Data_Value' columns to integers
df_cleaned.loc[:, ['Sample_Size', 'Data_Value']] = df_cleaned.loc[:, ['Sample_Size', 'Data_Value']].astype(int)

column_a = df_cleaned['Sample_Size']
column_b_percentages = df_cleaned['Data_Value']

# Convert percentage values to proportions
column_b_proportions = column_b_percentages / 100

# Multiply column A by the adjusted percentages
result = column_a * column_b_proportions

# Add the result as a new column in your DataFrame if needed
df_cleaned['Result_Column'] = result

# Filter Data for Method of Tabacco Usage to Sample_Size and Graph
ConsumptionType_df = df_cleaned[df_cleaned['TopicDesc'].isin(['E-Cigarette Use (Adults)',
                                       'Cigarette Use (Adults)', 
                                       'Smokeless Tobacco Use (Adults)'])]


ConsumptionType_data = ConsumptionType_df.groupby('TopicDesc')['Result_Column'].sum()

print(ConsumptionType_data)

# Plotting the bar chart
ConsumptionType_data.plot(kind='bar', color='orange')
plt.title('Comparison of Method of Tobacco Usage', fontweight = "bold")
plt.xlabel('Method of Tobacco Use', fontweight = "bold")
plt.ylabel('Total People using Tobacco', fontweight = "bold")
plt.xticks(rotation=45)
plt.show()

```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Assuming you have loaded your data into a DataFrame (smoke_data_ct)
smoke_data_ct = pd.read_csv("Behavioral_Risk_Factor_Data__Tobacco_Use__2017-Connecticut.csv")

selected_columns = ['LocationDesc', 'TopicDesc', 'MeasureDesc', 'Response', 'Sample_Size', 'Data_Value']
df_new = smoke_data_ct[selected_columns]

# Drop rows where 'Sample_Size' is NaN after conversion
df_cleaned = df_new.dropna(subset=['Sample_Size'])

# Convert 'Sample_Size' and 'Data_Value' columns to integers
df_cleaned.loc[:, ['Sample_Size', 'Data_Value']] = df_cleaned.loc[:, ['Sample_Size', 'Data_Value']].astype(int)

# Check if 'Data_Value' represents percentages (e.g., 50%)
if df_cleaned['Data_Value'].max() > 1:
    df_cleaned['Data_Value'] /= 100  # Convert percentages to proportions

# Multiply 'Sample_Size' by adjusted proportions and add the result as a new column
df_cleaned['Result_Column'] = df_cleaned['Sample_Size'] * df_cleaned['Data_Value']

# Plotting the pie chart
ConsumptionType_data = df_cleaned.groupby('TopicDesc')['Result_Column'].sum()
plt.pie(ConsumptionType_data, labels=ConsumptionType_data.index, autopct='%1.1f%%', startangle=140,
        colors=['lightcoral', 'skyblue', 'lightgreen'])
plt.title('Percentage Preference of Tobacco Use')
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have loaded your data into a DataFrame (smoke_data_ct)
smoke_data_ct = pd.read_csv("Behavioral_Risk_Factor_Data__Tobacco_Use__2017-Connecticut.csv")

selected_columns = ['LocationDesc', 'TopicDesc', 'MeasureDesc', 'Response', 'Sample_Size', 'Data_Value']
df_new = smoke_data_ct[selected_columns]

# Drop rows where 'Sample_Size' is NaN after conversion
df_cleaned = df_new.dropna(subset=['Sample_Size'])

# Convert 'Sample_Size' and 'Data_Value' columns to integers
df_cleaned.loc[:, ['Sample_Size', 'Data_Value']] = df_cleaned.loc[:, ['Sample_Size', 'Data_Value']].astype(int)

# Check if 'Data_Value' represents percentages (e.g., 50%)
if df_cleaned['Data_Value'].max() > 1:
    df_cleaned['Data_Value'] /= 100

# Multiply 'Sample_Size' by adjusted proportions and add the result as a new column
df_cleaned['Result_Column'] = df_cleaned['Sample_Size'] * df_cleaned['Data_Value']

# Group the data by the type of tobacco (smoking, smokeless, e-cigarette)
grouped_data = df_cleaned.groupby(['TopicDesc', 'Response'])

# Calculate the weighted average and merge the result back to the original DataFrame
df_cleaned = pd.merge(df_cleaned, grouped_data.apply(lambda x: np.average(x['Data_Value'], weights=x['Sample_Size'])).reset_index(name='Weighted_Average'), on=['TopicDesc', 'Response'])

# Step 3: Visualize the Data
plt.figure(figsize=(10, 6))

# Bar plot for different forms of tobacco
sns.barplot(x='TopicDesc', y='Weighted_Average', hue='Response', data=df_cleaned, palette=['lightcoral', 'skyblue', 'lightgreen'])

plt.title('Tobacco Usage in Adults (2017) - Connecticut')
plt.xlabel('Tobacco Type')
plt.ylabel('Weighted Average Percentage')
plt.show()
```
